```python
# 1 - Leer el archivo usando el módulo .csv
import csv

file_path = "13100504.csv"

with open(file_path, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    data = list(reader)

# Limpiar el encabezado (eliminar caracteres especiales si los hay)
header = [col.strip(' "\ufeff') for col in data[0]]
rows = data[1:]

# 2 - Contar cuántas filas y columnas hay (incluyendo o excluyendo encabezados)
num_rows = len(rows)
num_columns = len(header)

print(f"# 2 - Filas (sin encabezado): {num_rows}")
print(f"# 2 - Columnas: {num_columns}")

# 3 - Imprimir los primeros 5 registros usando estructuras de control (sin slicing)
print("# 3 - Primeros 5 registros:")
counter = 0
for row in rows:
    print(row)
    counter += 1
    if counter == 5:
        break

# 4 - Ver cuántos valores únicos hay en una columna específica (por ejemplo, 'Sex')
column_name = 'Sex'
col_index = header.index(column_name)
unique_values = set()

for row in rows:
    unique_values.add(row[col_index])

print(f"# 4 - Valores únicos en la columna '{column_name}': {len(unique_values)} -> {unique_values}")

# 5 - Calcular la media de una columna numérica (por ejemplo, 'VALUE')
value_index = header.index('VALUE')
values = []

for row in rows:
    try:
        val = float(row[value_index])
        values.append(val)
    except ValueError:
        continue  # Ignorar valores vacíos o no numéricos

mean_value = sum(values) / len(values) if values else 0
print(f"# 5 - Media de la columna 'VALUE': {mean_value:.2f}")

# 6 - Encontrar el valor máximo y mínimo en la columna 'VALUE'
max_value = max(values) if values else None
min_value = min(values) if values else None
print(f"# 6 - Valor máximo en 'VALUE': {max_value}")
print(f"# 6 - Valor mínimo en 'VALUE': {min_value}")

# 7 - Sorpresa: ¿Cuál es el año con mayor valor registrado en 'VALUE'?
ref_date_index = header.index('REF_DATE')
year_value_map = {}

for row in rows:
    try:
        year = row[ref_date_index]
        val = float(row[value_index])
        if year not in year_value_map or val > year_value_map[year]:
            year_value_map[year] = val
    except ValueError:
        continue

if year_value_map:
    best_year = max(year_value_map, key=year_value_map.get)
    print(f"# 7 - Año con mayor valor en 'VALUE': {best_year} ({year_value_map[best_year]})")
else:
    print("# 7 - No hay datos válidos para calcular el año con mayor valor.")
```
